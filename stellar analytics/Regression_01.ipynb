```python
# ═══════════════════════════════════════════════════════════════════════════════
# TASK B: REGRESSION - PLANETARY RADIUS PREDICTION (koi_prad)
# ═══════════════════════════════════════════════════════════════════════════════
# Section Header & Data Preparation

print("\n" + "="*80)
print("TASK B: REGRESSION - PREDICTING PLANETARY RADIUS (koi_prad)")
print("="*80 + "\n")

# Filter to only CONFIRMED exoplanets for regression task
df_regression = df[df['koi_disposition'] == 'CONFIRMED'].copy()

print(f"Regression Dataset Size: {len(df_regression):,} confirmed exoplanets")
print(f"Target: koi_prad (Planetary Radius in Earth Radii)")

# Check for missing values in target
print(f"\nMissing values in koi_prad: {df_regression['koi_prad'].isna().sum()}")

# Remove rows where koi_prad is missing (if any)
df_regression = df_regression.dropna(subset=['koi_prad'])

print(f"After removing missing koi_prad: {len(df_regression):,} samples\n")

# Target variable
y_reg = df_regression['koi_prad']

# Features - ONLY SELECT NUMERIC COLUMNS (excluding target and disposition)
# Get all numeric columns
numeric_cols = df_regression.select_dtypes(include=[np.number]).columns.tolist()

# Exclude target and disposition
feature_cols_reg = [col for col in numeric_cols 
                    if col not in ['koi_prad']]

X_reg = df_regression[feature_cols_reg]

print(f"Number of features: {len(feature_cols_reg)}")
print(f"Feature columns selected: {feature_cols_reg[:10]}...")  # Show first 10
print(f"\nTarget Statistics:")
print(f"  Mean Radius: {y_reg.mean():.2f} Earth Radii")
print(f"  Median Radius: {y_reg.median():.2f} Earth Radii")
print(f"  Min Radius: {y_reg.min():.2f} Earth Radii")
print(f"  Max Radius: {y_reg.max():.2f} Earth Radii")
print(f"  Std Dev: {y_reg.std():.2f} Earth Radii")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# Train-test split
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42
)

print(f"Regression Training set: {X_train_reg.shape[0]:,} samples")
print(f"Regression Test set: {X_test_reg.shape[0]:,} samples")

# Scaling
scaler_reg = StandardScaler()
X_train_reg_scaled = scaler_reg.fit_transform(X_train_reg)
X_test_reg_scaled = scaler_reg.transform(X_test_reg)

# Imputation
imputer_reg = SimpleImputer(strategy='mean')
X_train_reg_imputed = imputer_reg.fit_transform(X_train_reg_scaled)
X_test_reg_imputed = imputer_reg.transform(X_test_reg_scaled)

print("\nData preprocessing complete for regression task!")
```

```python
# Linear Regression Model

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Train Linear Regression
lr_reg_model = LinearRegression()
lr_reg_model.fit(X_train_reg_imputed, y_train_reg)
y_pred_lr_reg = lr_reg_model.predict(X_test_reg_imputed)

rmse_lr = np.sqrt(mean_squared_error(y_test_reg, y_pred_lr_reg))
mae_lr = mean_absolute_error(y_test_reg, y_pred_lr_reg)
r2_lr = r2_score(y_test_reg, y_pred_lr_reg)

print("Linear Regression Performance:")
print(f"  RMSE: {rmse_lr:.4f} Earth Radii")
print(f"  MAE:  {mae_lr:.4f} Earth Radii")
print(f"  R²:   {r2_lr:.4f}")
```

```python

# Train Random Forest Regressor

from sklearn.ensemble import RandomForestRegressor

# Train Random Forest Regressor
rf_reg_model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
rf_reg_model.fit(X_train_reg, y_train_reg)
y_pred_rf_reg = rf_reg_model.predict(X_test_reg)

rmse_rf = np.sqrt(mean_squared_error(y_test_reg, y_pred_rf_reg))
mae_rf = mean_absolute_error(y_test_reg, y_pred_rf_reg)
r2_rf = r2_score(y_test_reg, y_pred_rf_reg)

print("Random Forest Regressor Performance:")
print(f"  RMSE: {rmse_rf:.4f} Earth Radii")
print(f"  MAE:  {mae_rf:.4f} Earth Radii")
print(f"  R²:   {r2_rf:.4f}")
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# Train Gradient Boosting Regressor (using imputed data)
gb_reg_model = GradientBoostingRegressor(n_estimators=100, random_state=42)
gb_reg_model.fit(X_train_reg_imputed, y_train_reg)
y_pred_gb_reg = gb_reg_model.predict(X_test_reg_imputed)

rmse_gb = np.sqrt(mean_squared_error(y_test_reg, y_pred_gb_reg))
mae_gb = mean_absolute_error(y_test_reg, y_pred_gb_reg)
r2_gb = r2_score(y_test_reg, y_pred_gb_reg)

print("Gradient Boosting Regressor Performance:")
print(f"  RMSE: {rmse_gb:.4f} Earth Radii")
print(f"  MAE:  {mae_gb:.4f} Earth Radii")
print(f"  R²:   {r2_gb:.4f}")
```

```python
from sklearn.svm import SVR

# Train SVR (using imputed data - SVR requires scaled data)
svr_model = SVR(kernel='rbf', C=1.0, gamma='scale')
svr_model.fit(X_train_reg_imputed, y_train_reg)
y_pred_svr = svr_model.predict(X_test_reg_imputed)

rmse_svr = np.sqrt(mean_squared_error(y_test_reg, y_pred_svr))
mae_svr = mean_absolute_error(y_test_reg, y_pred_svr)
r2_svr = r2_score(y_test_reg, y_pred_svr)

print("Support Vector Regressor Performance:")
print(f"  RMSE: {rmse_svr:.4f} Earth Radii")
print(f"  MAE:  {mae_svr:.4f} Earth Radii")
print(f"  R²:   {r2_svr:.4f}")
```

```python
from sklearn.neural_network import MLPRegressor

# Train MLP Regressor (using imputed data)
mlp_reg_model = MLPRegressor(
    hidden_layer_sizes=(100, 50, 25),
    activation='relu',
    solver='adam',
    max_iter=500,
    random_state=42,
    early_stopping=True,
    validation_fraction=0.1,
    verbose=False
)
mlp_reg_model.fit(X_train_reg_imputed, y_train_reg)
y_pred_mlp_reg = mlp_reg_model.predict(X_test_reg_imputed)

rmse_mlp = np.sqrt(mean_squared_error(y_test_reg, y_pred_mlp_reg))
mae_mlp = mean_absolute_error(y_test_reg, y_pred_mlp_reg)
r2_mlp = r2_score(y_test_reg, y_pred_mlp_reg)

print("Neural Network (MLP) Regressor Performance:")
print(f"  RMSE: {rmse_mlp:.4f} Earth Radii")
print(f"  MAE:  {mae_mlp:.4f} Earth Radii")
print(f"  R²:   {r2_mlp:.4f}")
```

```python
from sklearn.neighbors import KNeighborsRegressor

# Train KNN Regressor (using imputed data - KNN requires scaled data)
knn_reg_model = KNeighborsRegressor(n_neighbors=5, weights='distance', n_jobs=-1)
knn_reg_model.fit(X_train_reg_imputed, y_train_reg)
y_pred_knn_reg = knn_reg_model.predict(X_test_reg_imputed)

rmse_knn = np.sqrt(mean_squared_error(y_test_reg, y_pred_knn_reg))
mae_knn = mean_absolute_error(y_test_reg, y_pred_knn_reg)
r2_knn = r2_score(y_test_reg, y_pred_knn_reg)

print("K-Nearest Neighbors Regressor Performance:")
print(f"  RMSE: {rmse_knn:.4f} Earth Radii")
print(f"  MAE:  {mae_knn:.4f} Earth Radii")
print(f"  R²:   {r2_knn:.4f}")
```

```python
# Install LightGBM if not already installed
import sys
!{sys.executable} -m pip install lightgbm --quiet

from lightgbm import LGBMRegressor

# Train LightGBM Regressor (using imputed data)
lgbm_reg_model = LGBMRegressor(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=7,
    num_leaves=31,
    random_state=42,
    n_jobs=-1,
    verbose=-1
)
lgbm_reg_model.fit(X_train_reg_imputed, y_train_reg)
y_pred_lgbm_reg = lgbm_reg_model.predict(X_test_reg_imputed)

rmse_lgbm = np.sqrt(mean_squared_error(y_test_reg, y_pred_lgbm_reg))
mae_lgbm = mean_absolute_error(y_test_reg, y_pred_lgbm_reg)
r2_lgbm = r2_score(y_test_reg, y_pred_lgbm_reg)

print("LightGBM Regressor Performance:")
print(f"  RMSE: {rmse_lgbm:.4f} Earth Radii")
print(f"  MAE:  {mae_lgbm:.4f} Earth Radii")
print(f"  R²:   {r2_lgbm:.4f}")
```

```python
from sklearn.ensemble import VotingRegressor

print("Training Voting Regressor Ensemble...")

# Retrain tree models on imputed data for ensemble consistency
rf_reg_ensemble = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
rf_reg_ensemble.fit(X_train_reg_imputed, y_train_reg)

gb_reg_ensemble = GradientBoostingRegressor(n_estimators=100, random_state=42)
gb_reg_ensemble.fit(X_train_reg_imputed, y_train_reg)

lgbm_reg_ensemble = LGBMRegressor(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=7,
    num_leaves=31,
    random_state=42,
    n_jobs=-1,
    verbose=-1
)
lgbm_reg_ensemble.fit(X_train_reg_imputed, y_train_reg)

# Create Voting Regressor
voting_reg_model = VotingRegressor(
    estimators=[
        ('lr', lr_reg_model),
        ('rf', rf_reg_ensemble),
        ('gb', gb_reg_ensemble),
        ('svr', svr_model),
        ('mlp', mlp_reg_model),
        ('lgbm', lgbm_reg_ensemble)
    ]
)

voting_reg_model.fit(X_train_reg_imputed, y_train_reg)
y_pred_voting_reg = voting_reg_model.predict(X_test_reg_imputed)

rmse_voting = np.sqrt(mean_squared_error(y_test_reg, y_pred_voting_reg))
mae_voting = mean_absolute_error(y_test_reg, y_pred_voting_reg)
r2_voting = r2_score(y_test_reg, y_pred_voting_reg)

print("Voting Regressor (Ensemble) Performance:")
print(f"  RMSE: {rmse_voting:.4f} Earth Radii")
print(f"  MAE:  {mae_voting:.4f} Earth Radii")
print(f"  R²:   {r2_voting:.4f}")
```

```python
# Store regression results
regression_results = {
    'Linear Regression': {
        'predictions': y_pred_lr_reg,
        'rmse': rmse_lr,
        'mae': mae_lr,
        'r2': r2_lr
    },
    'Random Forest': {
        'predictions': y_pred_rf_reg,
        'rmse': rmse_rf,
        'mae': mae_rf,
        'r2': r2_rf
    },
    'Gradient Boosting': {
        'predictions': y_pred_gb_reg,
        'rmse': rmse_gb,
        'mae': mae_gb,
        'r2': r2_gb
    },
    'SVR': {
        'predictions': y_pred_svr,
        'rmse': rmse_svr,
        'mae': mae_svr,
        'r2': r2_svr
    },
    'Neural Network': {
        'predictions': y_pred_mlp_reg,
        'rmse': rmse_mlp,
        'mae': mae_mlp,
        'r2': r2_mlp
    },
    'K-Nearest Neighbors': {
        'predictions': y_pred_knn_reg,
        'rmse': rmse_knn,
        'mae': mae_knn,
        'r2': r2_knn
    },
    'LightGBM': {
        'predictions': y_pred_lgbm_reg,
        'rmse': rmse_lgbm,
        'mae': mae_lgbm,
        'r2': r2_lgbm
    },
    'Voting Ensemble': {
        'predictions': y_pred_voting_reg,
        'rmse': rmse_voting,
        'mae': mae_voting,
        'r2': r2_voting
    }
}

# Print summary table
print("\n" + "="*85)
print("REGRESSION MODEL PERFORMANCE SUMMARY - ALL 8 ALGORITHMS")
print("="*85)
print(f"{'Model':<22} {'RMSE (Earth Radii)':<20} {'MAE (Earth Radii)':<20} {'R² Score':<15}")
print("-"*85)
for model_name, results in regression_results.items():
    print(f"{model_name:<22} {results['rmse']:<20.4f} {results['mae']:<20.4f} {results['r2']:<15.4f}")
print("="*85)

# Find best models
best_rmse_model = min(regression_results.items(), key=lambda x: x[1]['rmse'])
best_mae_model = min(regression_results.items(), key=lambda x: x[1]['mae'])
best_r2_model = max(regression_results.items(), key=lambda x: x[1]['r2'])

print("\nBEST MODELS FOR EACH METRIC:")
print("-"*60)
print(f"  Lowest RMSE: {best_rmse_model[0]} ({best_rmse_model[1]['rmse']:.4f} Earth Radii)")
print(f"  Lowest MAE:  {best_mae_model[0]} ({best_mae_model[1]['mae']:.4f} Earth Radii)")
print(f"  Highest R²:  {best_r2_model[0]} ({best_r2_model[1]['r2']:.4f})")
print("="*85)
```
