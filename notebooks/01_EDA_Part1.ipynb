```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

# Load data
df = pd.read_csv('../data/cumulative.csv')

# Basic info
print(f"Dataset shape: {df.shape}")
print(f"\nColumn count: {len(df.columns)}")
print(f"\nFirst few rows:")
df.head()
```

```python
print("All columns:")
for i, col in enumerate(df.columns, 1):
    print(f"{i:2d}. {col}")
```

```python
print("Data types:")
df.dtypes
```

```python
print("=== TARGET VARIABLE: koi_disposition ===")
print(df['koi_disposition'].value_counts()) 
print("\n% Distribution:")
print(df['koi_disposition'].value_counts(normalize=True) * 100)

print("\n=== TARGET VARIABLE: koi_prad ===")
print(df['koi_prad'].describe())# We can find outliers in the data by this .
```

```python
# Calculate missing values
missing = df.isnull().sum()
missing_percent = (missing / len(df)) * 100

# Create DataFrame
missing_df = pd.DataFrame({
    'Column': df.columns,
    'Missing_Count': missing.values,
    'Percentage': missing_percent.values
}).sort_values('Percentage', ascending=False)

# Show only columns with missing values
missing_with_values = missing_df[missing_df['Missing_Count'] > 0]

print(f"Columns with missing values: {len(missing_with_values)}")
print("\nTop 20 columns with most missing data:")
missing_with_values.head(20)
```

```python
# Save to docs folder
missing_df.to_csv('../docs/missing_values_report.csv', index=False)
print("✅ Report saved to docs/missing_values_report.csv")
```
